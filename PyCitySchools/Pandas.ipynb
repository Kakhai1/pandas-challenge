```python
from pathlib import Path
import pandas as pd
```

```python
#Define paths for csv files
file = Path("Resources/schools_complete.csv")
file2 = Path("Resources/students_complete.csv")
```

```python
#Define Dataframes
dfschool = pd.read_csv(file)
dfstudent = pd.read_csv(file2)
```

```python
#Merge dataframes into working dataframe
mergedf = pd.merge(dfschool,dfstudent,on = "school_name")
mergedf
```

```python
mergedf["year"].min()
```

```python
mergedf["year"].max()
```

```python
#Count number of unique schools
uniqueschool = mergedf["school_name"].nunique()
uniqueschool
```

```python
#Count Students
totalstudent = mergedf["Student ID"].count()
totalstudent
```

```python
#Sum of total budget
totalbudget = dfschool["budget"].sum()
totalbudget
```

```python
#Calculate average scores
avgmath = mergedf["maths_score"].mean()
avgmath
```

```python
avgreading = mergedf["reading_score"].mean()
avgreading
```

```python
#Create a list of boolean True or False (1 or 0) and average that list to find % passing
passingmath = (mergedf["maths_score"] >= 50).mean()*100
passingmath
```

```python
#Same as Math
passingreading = (mergedf["reading_score"] >= 50).mean()*100
passingreading
```

```python
#Same as Math and Reading, but with two conditions instead of 1
passingoverall = ((mergedf["maths_score"] >= 50) & (mergedf["reading_score"] >= 50)).mean() * 100
passingoverall
```

```python
#Consolidate information into  data frame
LGAdf = pd.DataFrame({
    "Total unique schools":uniqueschool,
    "Total students":totalstudent,
    "Total Budget":totalbudget,
    "Average maths score":avgmath,
    "Average reading score":avgreading,
    "% Passing Math":passingmath,
    "% Passing Reading":passingreading,
    "% Passing Overall":passingoverall,
},
    index = [""]
)
LGAdf
```

```python
#Create variable for next few calculations
schoolgroup = mergedf.groupby("school_name")
```

```python
totalgroupstudents = schoolgroup["student_name"].count()
totalgroupstudents
```

```python
#Use first to count once per school name
groupschooltype = schoolgroup["type"].first()
groupschooltype
```

```python
#Use first to count once per school name
groupschoolbudget = schoolgroup["budget"].first()
groupschoolbudget
```

```python
perstudentbudget = groupschoolbudget/totalgroupstudents
perstudentbudget
```

```python
groupmathavg = schoolgroup["maths_score"].mean()
groupmathavg
```

```python
groupreadingavg = schoolgroup["reading_score"].mean()
groupreadingavg
```

```python
passingmathgroup = (mergedf[mergedf["maths_score"] >= 50].groupby("school_name")["student_name"].count() 
                    / totalgroupstudents) * 100
passingmathgroup
```

```python
passingreadgroup = (mergedf[mergedf["reading_score"] >= 50].groupby("school_name")["student_name"].count() 
                    / totalgroupstudents) * 100
passingreadgroup
```

```python
passingoverallgroup = (mergedf[(mergedf['maths_score'] >= 50) & (mergedf['reading_score'] >= 50)]
                              .groupby('school_name')['student_name'].count() / totalgroupstudents) * 100
passingoverallgroup
```

```python
schoolsummary = pd.DataFrame({
    "School Type":groupschooltype,
    "Total Students":totalgroupstudents,
    "Total School Budget":groupschoolbudget,
    "Per Student Budget":perstudentbudget,
    "Average Math Score":groupmathavg,
    "Average Reading Score":groupreadingavg,
    "% Passing Math":passingmathgroup,
    "% Passing Reading":passingreadgroup,
    "% Passing Overall":passingoverallgroup,
}
)
schoolsummary
```

```python
topschools = schoolsummary.sort_values("% Passing Overall", ascending = False)
topschools.head()
```

```python
bottomschools = schoolsummary.sort_values("% Passing Overall", ascending = True)
bottomschools.head()
```

```python
#Dual filter groupby to score
studentfiltermath = (mergedf.groupby(["school_name","year"])["maths_score"].mean().reset_index())
studentfiltermath 
```

```python
studentfilterread = (mergedf.groupby(["school_name","year"])["reading_score"].mean().reset_index())
studentfilterread
```

```python
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
cutschool = pd.cut(schoolsummary["Per Student Budget"],spending_bins,labels=labels, include_lowest = True)

schoolsummary["Spending Ranges (Per Student)"] = cutschool
schoolsummary
```

```python
#preparing variables for DataFrame
spending_math_scores = schoolsummary.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = schoolsummary.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = schoolsummary.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = schoolsummary.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = schoolsummary.groupby(["Spending Ranges (Per Student)"])["% Passing Overall"].mean()
```

```python
spending_summary = pd.DataFrame({
    "Average maths score":spending_math_scores,
    "Average reading score":spending_reading_scores,
    "% passing maths":spending_passing_math,
    "% passing reading":spending_passing_reading,
    "% overall passing":overall_passing_spending
})
spending_summary
```

```python
size_bins = [0, 1000, 2000, 5000]
labels2 = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
cutschoolsize = pd.cut(schoolsummary["Total Students"],size_bins,labels = labels2,include_lowest = True)
#Creating new column for school size
schoolsummary["School size"] = cutschoolsize

#Preparing variables for DataFrame using new column
size_math_scores = schoolsummary.groupby(["School size"])["Average Math Score"].mean()
size_reading_scores = schoolsummary.groupby(["School size"])["Average Reading Score"].mean()
size_passing_math = schoolsummary.groupby(["School size"])["% Passing Math"].mean()
size_passing_reading = schoolsummary.groupby(["School size"])["% Passing Reading"].mean()
overall_passing_size = schoolsummary.groupby(["School size"])["% Passing Overall"].mean()
```

```python
size_summary = pd.DataFrame({
    "Average maths score":size_math_scores,
    "Average reading score":size_reading_scores,
    "% passing maths":size_passing_math,
    "% passing reading":size_passing_reading,
    "% overall passing":overall_passing_size
})
size_summary
```

```python
#No cut needed
type_math_scores = schoolsummary.groupby(["School Type"])["Average Math Score"].mean()
type_reading_scores = schoolsummary.groupby(["School Type"])["Average Reading Score"].mean()
type_passing_math = schoolsummary.groupby(["School Type"])["% Passing Math"].mean()
type_passing_reading = schoolsummary.groupby(["School Type"])["% Passing Reading"].mean()
overall_passing_type = schoolsummary.groupby(["School Type"])["% Passing Overall"].mean()

type_summary = pd.DataFrame({
    "Average maths score":type_math_scores,
    "Average reading score":type_reading_scores,
    "% passing maths":type_passing_math,
    "% passing reading":type_passing_reading,
    "% overall passing":overall_passing_type
})
type_summary
```
